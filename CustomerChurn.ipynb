```python

from sklearn.preprocessing import LabelEncoder
import numpy as np

# Loop through each column in the dataset
for column in dataset.columns:
    # Check if the column's data type is not a number
    if dataset[column].dtype == np.number:
        # If it's a number, skip this iteration
        continue
    # If it's not a number, apply LabelEncoder to the column
    dataset[column] = LabelEncoder().fit_transform(dataset[column])
```

Feature Scaling

```python

from sklearn.preprocessing import StandardScaler

# Instantiate the scaler
scaler = StandardScaler()

# Fit and transform the dataset
X = scaler.fit_transform(X)
```

Splitting the dataset into training and test sets

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size =0.2,random_state =0)
```

```python
print(X_train)
```

BUILDING THE LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
classifier =LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

PREDICTIONS

```python
y_predict=classifier.predict(X_test)
```

Predicting the test results

```python
np.set_printoptions(precision=2)
```

```python



# Convert y_test (which is a pandas Series) to a NumPy array before reshaping
result = np.concatenate((y_predict.reshape(len(y_predict), 1),
                         y_test.values.reshape(len(y_test), 1)), axis=1)

# Print the result
print(result)
```

Model Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm =confusion_matrix(y_test,y_predict)
accuracy =accuracy_score(y_test,y_predict)
```
